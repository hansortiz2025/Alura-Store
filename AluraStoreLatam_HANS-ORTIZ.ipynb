# 3. Calificación promedio de la tienda


```python
print("CLASIFICACIÓN PROMEDIO")
calificaciones_promedio = []

for i in range(len(tiendas)):
    # Calculamos el promedio de la columna Calificación
    promedio_calif = tiendas[i]['Calificación'].mean()
    calificaciones_promedio.append(promedio_calif)
    print(f"{nombres_tiendas[i]}: {promedio_calif:.2f} estrellas")
```

# 4. Productos más y menos vendidos

```python
print("PRODUCTOS MÁS Y MENOS VENDIDOS")

for i in range(len(tiendas)):
    # Contamos la frecuencia de cada producto
    conteo_productos = tiendas[i]['Producto'].value_counts()

    # El índice 0 es el que más se repite, el índice -1 es el que menos
    mas_vendido = conteo_productos.index[0]
    menos_vendido = conteo_productos.index[-1]

    print(f"\n{nombres_tiendas[i]}:")
    print(f"  Más vendido: {mas_vendido} ({conteo_productos.iloc[0]} unidades)")
    print(f"  Menos vendido: {menos_vendido} ({conteo_productos.iloc[-1]} unidades)")
```

# 5. Envío promedio por tienda
